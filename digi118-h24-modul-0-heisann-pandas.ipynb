## 3b. `loc[]`
En annen m친te 친 velge data p친 bruker metoden [`loc[]`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.loc.html). Denne metoden **lokaliserer** og trekker ut rader og kolonner innenfor datarammen som vi er interessert i. Vi spesifiserer v친r foresp칮rsel innenfor `[]`-parentesene.

`loc[]` kan ta enkle foresp칮rsler, som 친 finne og returnere informasjon fra en bestemt rad i v친r dataramme. For 친 returnere en rad, gir vi `loc[]` indeksen til den raden.

```python
# Finn radindeks 3. Dette returnerer en serie.
df.loc[3]
```

`loc[]` kan ogs친 ta p친 seg mer komplekse foresp칮rsler, for eksempel 친 finne og returnere flere rader eller kolonner innenfor datarammen v친r. Vi kan be `loc[]` om 친 returnere enten en liste med rader eller deler av rader.

*Trenger du en oppfriskning p친 lister eller skj칝ring? Se [DIGI118 Modul 0: Heisann, Python!](https://www.kaggle.com/code/lauragarrison/digi118-h24-modul-0-heisann-pandas)*

```python
# Finn radindeksene 3 og 7. Dette returnerer en dataramme.
# Legg merke til de doble parentesene her
df.loc[[3, 7]]
```

```python
# Finn rad 3 til 7 ved hjelp av skiver. Dette returnerer en dataramme.
# OBS: I motsetning til vanlige pytonskiver er b친de start- og stoppindeksen inkludert!
df.loc[3:7]
```

`loc[]` kan ogs친 returnere rader eller kolonner bare om de passer til en bestemt betingelse.

La oss si at vi bare vil se informasjon om byen Seattle. V친r foresp칮rsel vil v칝re 친 finne alle rader der verdiene i kolonnen `df["city"]` tilsvarer (`==`) navnet p친 byen,`"Seattle"`.

```python
df.loc[df["city"] == "Seattle"]
```

## 3c. `iloc[]`
En tredje m친te 친 velge data p친 er 친 bruke metoden [`iloc[]`](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iloc.html). Denne metoden kan gj칮re mer spesifikke valg enn de to andre vi har beskrevet s친 langt. *I dette kurset vil vi ikke bruke denne metoden i detalj, og vi har tatt med et enkelt eksempel bare for 친 vise hvordan den brukes.*

`iloc[]` bruker indeksene til radene og kolonnene for 친 finne verdier i en dataramme. Vi kan skrive en foresp칮rsel slik: `iloc[row, column]`.

For eksempel, hvis vi 칮nsker 친 vite nedb칮ren til Seattle i desember, gir vi `iloc[]` radindeksen til Seattle i desember og kolonneindeksen for nedb칮r. Vi kan se fra kodecellen over at dette er rad 2 og kolonne 2, s친 bruker vi denne informasjonen i foresp칮rselen.

```python
df.iloc[2, 2]
```

Hvis vi 칮nsker 친 trekke ut en rekke rader og kolonner, kan vi gi `iloc[]` skiver i stedet for indekser. I motsetning til `loc[]` f칮lger `iloc[]` reglene for skj칝ring i pyton der stoppindeksen ikke er inkludert i utdataene.

```python
df.iloc[ 1:5 , 1:3 ] 
```

# 5. Grunnleggende dataanalyse
Etter 친 ha lest inn dataene v친re, kan vi bruke `pandas`-funksjoner til 친 beregne **beskrivende statistikk** som kvantitativt beskriver og oppsummerer datasettet v친rt. Eksempler p친 beskrivende statistikk inkluderer:
- `count()`: det totale antallet elementer i et datasett som ikke er null eller mangler verdier
- `mean()`: gjennomsnittet av verdier i et sett med data
- `median()`: den midterste verdien av et datasett n친r det er sortert fra minst til st칮rst
- **maksimum** eller 'maks()'-verdien i et datasett
- `sum()`: det totale bel칮pet som er resultatet av 친 legge til to eller flere verdier i et datasett

La oss stille dataene v친re noen sp칮rsm친l. F.eks, hvor mange forskjellige aldre er registrert i dette datasettet?

```python
titanic["Age"].count()
```

Vi ser at det var registrert 714 forskjellige aldre. N친 lurer jeg p친 hva gjennomsnittsalderen og gjennomsnittlig billettpris var for alle passasjerer?

```python
titanic[["Age", "Fare"]].mean()
```

Og hva var medianalderen for passasjerer p친 Titanic?

```python
titanic["Age"].median()
```

Medianalderen var 28 친r gammel. Jeg lurer n친 hvor gammel var den eldste passasjeren om bord p친 Titanic?

```python
titanic["Age"].max()
```

Dette er interessant, og jeg vil vite **hvem** som var denne eldste passasjeren om bord p친 Titanic?

Vi kan bruke `idxmax()` for 친 finne radindeksen til denne eldste passasjeren, og bruke `loc[]` for 친 finne denne raden.

```python
titanic.loc[titanic["Age"].idxmax()]
```

Det var Mr. Algernon Henry Wilson Barkworth (for et navn!) som var den eldste passasjeren ombord, og han var i 1. klasse. Vi ser ogs친 at han overlevd. Hvem var den eldste passasjeren om bord p친 Titanic som *ikke* overlevde? Vi kan finne dette i to trinn. Vi kunne lenke dette sammen, men for klarhetens skyld viser vi dem hver for seg.

```python
# Finn f칮rst alle passasjerer som overlevde Titanic. I dette datasettet indikerer 0 en passasjer som ikke overlevde.
survivors = titanic.loc[titanic["Survived"] == 0]

# Finn den eldste passasjeren i disse passasjerene.
survivors.loc[survivors["Age"].idxmax()]
```

Her ser vi hvordan billetttypen p친virket overlevelsen. Johan Svensson var den eldste ikke-overlevende, og han var i 3. klasse.

N친r vi tenker p친 billettprisen, hvor mye penger ble samlet inn for Titanic-billettpriser? Vi skal lagre dette i en variabel for 친 bruke senere

```python
total_fare = titanic["Fare"].sum()
total_fare
```

Hvor mange prosent av dette ble nettopp hentet fra 1. klasse passasjerer?

```python
# What percentage of that fare was collected from first class passengers?
first_fare = titanic.loc[titanic["Pclass"] == 1]["Fare"].sum()
first_fare / total_fare * 100
```

Ca. 63% av totalt billetprisen kom fra 1. klasse. 

I eksemplene ovenfor fant vi ofte et **delsett** av 춺titanic췉-datarammen f칮rst f칮r vi brukte andre metoder p친 dataene.

Som nevnt tidligere, er `loc[]` en av de nyttige metodene vi kan bruke for 친 trekke ut spesifikke rader og kolonner fra en dataramme. Vi lagrer denne mindre datarammen i en ny variabel for 친 jobbe med.

Etter 친 ha opprettet et undersett av data, kan vi bruke `info()` for 친 sikre at vi har trukket ut dataene riktig.

F.eks, la oss for eksempel se p친 datasettet med bare attributtene `"Survived`", `"Pclass"`, `"Age"` og `"Sex"` inkludert.

```python
titanic_subset = titanic.loc[:, ["Survived", "Pclass", "Age", "Sex"]]
titanic_subset.info()
```

En av fordelene med 친 jobbe med et undersett av det st칮rre datasettet er at du har f칝rre un칮dvendige beregninger 친 utf칮re, og koden din vil kj칮re raskere.
