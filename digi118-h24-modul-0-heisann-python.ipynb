# 3. Variabler
En variabel er en referanse til en verdi. Det kan tenkes på som en "lagringsbeholder" for en verdi, og denne verdien **kan** endres.

Vi kan navngi variablene hva vi vil, men de må begynne med en bokstav og kan ikke inneholde spesialtegn (bortsett fra `_`). Det er heller ikke en god idé å navngi en variabel etter et Python innebygd nøkkelord eller funksjonsnavn. Du kan lese mer om [gode variabelnavn her](https://emnearkiv.ii.uib.no/inf100/23V/notat/ekstra_notes_on_variables/#Gode-variabelnavn).

Vi kan tilordne en verdi til en variabel ved å bruke tilordningsoperasjonen, `=`. La oss lage en ny variabel, `my_number`, og tilordne den verdien `2`:

````python
# Tilordne verdier til nye variabler
my_number = 2

my_number
````

Det er to måter vi kan oppdatere verdien av en variabel. En måte er å **opprette en ny variabel** for å lagre denne nye verdien. På denne måten kan vi referere til både den nye verdien og den gamle verdien.

````python
my_new_number = my_number + 1

my_new_number
````

En annen måte er å **overskrive vår gamle variabel** med den nye verdien. Dette betyr at det ikke finnes referansen til den gamle verdien.

````python
# Overskrive verdi i gammel variabel
my_number = my_number + 1

my_number
````

Selvfølgelig kunne vi ha overført verdien av vår gamle variabel til en ny variabel for å lagre den til senere, men dette må gjøres **før** vi overskriver den. **Rekkefølgen er viktig!**

# 4. `List`
Nå som vi har introdusert variabler, er det lettere å snakke om lister, som er den siste datatypen vi skal diskutere i dette kurset.

En `list` er en spesiell datatype som brukes til å lagre flere verdier. Lister kan opprettes ved å plassere verdier (atskilt med komma) innenfor firkantede parenteser:

````python
my_list = ["apple", "orange", "cherry"]
````

Elementer i en liste har en **definert rekkefølge**. Dette betyr at vi kan referere til elementer etter deres plassering i sekvensen, også kjent som deres **indeks**.

I Python vil det første elementet i en sekvens starte med en indeks på `0`, det andre elementet med en indeks på `1`, osv. Vi kan få tilgang til en sekvens ved å bruke dens indeks ved å bruke hakeparentesnotasjonen `[]`, også kjent som indekseringsoperatorene.

````python
# Vi må kalle den første indeksen i my_list for å skrive ut den andre listelementen, "orange"
my_list[1]
````

Elementer i en listen kan inholde **forskjellige datatypene**. 

````python
["apple", 1, 5.34, False, True]
````

Siden lister er indeksert, elementer i en list kan ha **samme verdi**.

````python
["apple", "apple"]
````

Elementer i en liste kan også **skiftes**, noe som betyr at vi kan oppdatere, legge til og fjerne elementer i en liste etter at listen er opprettet. Vi kan gjøre dette med noen av Pythons innebygde funksjoner, som vi skal berøre i neste delen av notatboken, men vi kan også redigere lister ved hjelp av skjæring (slicing på engelsk).

I Python kan du dele opp en sekvens, for eksempel en `string` eller en `list`, for å returnere et spesifikt delsett av elementer uten å endre den opprinnelige. Dette gjør skjæring til et nyttig verktøy for datamanipulering og utvinning.

For å skjære noe, må vi spesifisere hvor vi skal starte skjæringen, samt hvor vi skal slutte. Vi skriver dette ved å bruke en hakeparentes med kolon som skiller startindeksen, sluttindeksen og trinnet til skiven (f.eks. skjær bare annenhver gjenstand): `list[0:3:2]`.

**OBS:** Elementet i sluttindeksen er **ikke** inkludert i snittet!

````python
fruits = ["apple", "orange", "banana", "grapes"]
````

````python
# bruk skjæring for å oppdatere et listeelement
fruits[0] = "kiwi"

fruits
````

````python
# Lag en skive med "orange" og "banana". Merk at "grapes" med indeks 3 ikke er inkludert.
my_fruits = fruits[1:3]

my_fruits
````

````python
# Men dette vil skape en skive med "orange", "banana" og "grapes".
my_other_fruits = fruits[1:4]

my_other_fruits
````

````python
# Og dette skaper et stykke med hvert annet element i listen
fruits[0:5:2]
````

Om vi ønsker å skjære hele veien fra starten av listen, kan vi utelate startindeksen. På samme måte, hvis vi ønsker å dele hele veien til slutten av listen, kan vi utelate sluttindeksen.

````python
# Dette lager et stykke fra starten av listen til "banana"
fruits[ :3]
````

````python
# Mens dette skaper et stykke fra "banana" til slutten av listen
fruits[2: ]
````

````python
# Og dette skaper en del av hele listen
fruits[ : ]
````

# 5. Funksjoner og metoder
Funksjoner og metoder er kommandoer som du gir datamaskinen for å utføre en bestemt oppgave. Disse kommandoene er enten innebygd i Python (dvs. metoder), eller kan være en egendefinert sekvens av kommandoer.

Generelt er det en god idé å skrive funksjoner for vanlige eller gjentatte oppgaver. I stedet for å skrive den samme koden om og om igjen for forskjellige innganger, kan vi kalle en funksjon for å gjenbruke den samme kodeblokken.

## 5a. Funksjoner 
En egendefinert funksjon er definert ved å bruke nøkkelordet `def` og et funksjonsnavn etter eget valg. Deretter kommer oppgaven(ene) vi vil at funksjonen vår skal utføre.

Vanligvis vil vi at funksjonen vår skal ta noen input i form av ***argumenter***, som er verdier som sendes innenfor parentesen til funksjonen. Deretter vil vi at funksjonen skal behandle denne inngangen på en eller annen måte, og ***returnere*** en utgang.

**OBS:** *I dette kurset vil du ikke bli bedt om å skrive funksjoner selv, men heller bli utstyrt med "hjelper"-funksjoner for å fullføre oppgaven.*

````python
# Vi ønsker å skrive en funksjon som kan legge til to tall og returnere summen
# Vi definerer denne funksjonen som "add_numbers()"
# Vi tar to tall, num1 og num2, som input og returnerer num3, summen av inngangene

def add_numbers(num1, num2):
    num3 = num1 + num2
    
    return num3
````

Deretter **kaller vi funksjonen** ved å skrive funksjonsnavnet og sende argumenter innenfor dens parentes:

````python
# Vi ønsker å legge tallene 10 og 20 sammen
add_numbers(10, 20)
````

## 5b. Metoder 
En metode er en spesiell type funksjon som er assosiert med et datatype i Python, for eksempel en `list` eller en `string`. Python har et sett med innebygde metoder for hvert av disse datatypene. Du kan finne en [utfyllende liste over de forskjellige metodene her](https://www.w3schools.com/python/python_reference.asp).

Det er flere nyttige metoder for å legge til eller fjerne elementer fra lister. Her introduserer vi bare `append()`, som legger til et element(er) på slutten av listen din, og `remove()`, som fjerner et element du spesifiserer fra listen din. For å se noen av de andre tingene du kan gjøre med lister, er det mange ressurser tilgjengelig, for eksempel [Python Lists and List Manipulation, Explained](https://builtin.com/data-science/python-list).

Vi vil også vise deg en nyttig metode for å rydde opp i en strengs formatering, og hvordan du kaller flere metoder på en enkelt linje av koden.  

### Nyttige `list`-metoder -- `append()` og `remove()`

````python
# legg til elementet på slutten av listen
my_list = ["apple", "orange", "cherry"]
my_list.append("banana")

my_list
````

````python
# ta vekk elementet fra listen
my_list = ["apple", "orange", "cherry"]
my_list.remove("apple")

my_list
````

### Rydding av strenger
`strip()` er en nyttig metode som fjerner ekstra tegn fra begynnelsen og slutten av en streng:

````python
# strip() fjerner innledende og etterfølgende mellomrom som standard
second_string = "         Welcome to DIGI118                   "

second_string.strip()
````

Til slutt er det mulig å bruke metoder på en variabel etter hverandre. Vi kaller dette **kjetting**. Hvis vi ønsker å fjerne mellomrom i tekst og/eller endre noen av ordene, kan vi lenke `strip()` og `replace()` sammen i én kodelinje:

````python
text = ",,,,,,,Welcome to DIGI118!,,,,,"
text_changes = text.strip(",").replace("DIGI118", "Introduction to Data Visualization!")

text_changes
````
