En variabel er en referanse til en verdi. Det kan tenkes pÃ¥ som en "lagringsbeholder" for en verdi, og denne verdien **kan** endres.

Vi kan navngi variablene hva vi vil, men de mÃ¥ begynne med en bokstav og kan ikke inneholde spesialtegn (bortsett fra `_`). Vi kan tilordne en verdi til en variabel ved Ã¥ bruke tilordningsoperasjonen, `=`. La oss lage en ny variabel, `my_number`, og tilordne den verdien `2`:

````python
# Tilordne verdier til nye variabler
my_number = 2

my_number
````

Det er to mÃ¥ter vi kan oppdatere verdien av en variabel. En mÃ¥te er Ã¥ **opprette en ny variabel** for Ã¥ lagre denne nye verdien. PÃ¥ denne mÃ¥ten kan vi referere til bÃ¥de den nye verdien og den gamle verdien.

````python
my_new_number = my_number + 1

my_new_number
````

# 4. `List`
NÃ¥ som vi har introdusert variabler, er det lettere Ã¥ snakke om lister, som er den siste datatypen vi skal diskutere i dette kurset.

En `list` er en spesiell datatype som brukes til Ã¥ lagre flere verdier. Lister kan opprettes ved Ã¥ plassere verdier (atskilt med komma) innenfor firkantede parenteser:

````python
my_list = ["apple", "orange", "cherry"]
````

Elementer i en liste har en **definert rekkefÃ¸lge**. Dette betyr at vi kan referere til elementer etter deres plassering i sekvensen, ogsÃ¥ kjent som deres **indeks**.

I Python vil det fÃ¸rste elementet i en sekvens starte med en indeks pÃ¥ `0`, det andre elementet med en indeks pÃ¥ `1`, osv. Vi kan fÃ¥ tilgang til en sekvens ved Ã¥ bruke dens indeks ved Ã¥ bruke hakeparentesnotasjonen `[]`, ogsÃ¥ kjent som indekseringsoperatorene.

````python
# Vi mÃ¥ kalle den fÃ¸rste indeksen i my_list for Ã¥ skrive ut den andre listelementen, "orange"
my_list[1]
````

Elementer i en listen kan inholde **forskjellige datatypene**. 

````python
["apple", 1, 5.34, False, True]
````

Siden lister er indeksert, elementer i en list kan ha **samme verdi**.

````python
["apple", "apple"]
````

Elementer i en liste kan ogsÃ¥ **skiftes**, noe som betyr at vi kan oppdatere, legge til og fjerne elementer i en liste etter at listen er opprettet. Vi kan gjÃ¸re dette med noen av Pythons innebygde funksjoner, som vi skal berÃ¸re i neste delen av notatboken, men vi kan ogsÃ¥ redigere lister ved hjelp av skjÃ¦ring (slicing pÃ¥ engelsk).

I Python kan du dele opp en sekvens, for eksempel en `string` eller en `list`, for Ã¥ returnere et spesifikt delsett av elementer uten Ã¥ endre den opprinnelige. Dette gjÃ¸r skjÃ¦ring til et nyttig verktÃ¸y for datamanipulering og utvinning.

For Ã¥ skjÃ¦re noe, mÃ¥ vi spesifisere hvor vi skal starte skjÃ¦ringen, samt hvor vi skal slutte. Vi skriver dette ved Ã¥ bruke en hakeparentes med kolon som skiller startindeksen, sluttindeksen og trinnet til skiven (f.eks. skjÃ¦r bare annenhver gjenstand): `list[0:3:2]`.

**OBS:** Elementet i sluttindeksen er **ikke** inkludert i snittet!

````python
fruits = ["apple", "orange", "banana", "grapes"]
````

````python
# bruk skjÃ¦ring for Ã¥ oppdatere et listeelement
fruits[0] = "kiwi"

fruits
````

````python
# Lag en skive med "orange" og "banana". Merk at "grapes" med indeks 3 ikke er inkludert.
my_fruits = fruits[1:3]

my_fruits
````

````python
# Men dette vil skape en skive med "orange", "banana" og "grapes".
my_other_fruits = fruits[1:4]

my_other_fruits
````

````python
# Og dette skaper et stykke med hvert annet element i listen
fruits[0:5:2]
````

Om vi Ã¸nsker Ã¥ skjÃ¦re hele veien fra starten av listen, kan vi utelate startindeksen. PÃ¥ samme mÃ¥te, hvis vi Ã¸nsker Ã¥ dele hele veien til slutten av listen, kan vi utelate sluttindeksen.

````python
# Dette lager et stykke fra starten av listen til "banana"
fruits[ :3]
````

````python
# Mens dette skaper et stykke fra "banana" til slutten av listen
fruits[2: ]
````

````python
# Og dette skaper en del av hele listen
fruits[ : ]
````
